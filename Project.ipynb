**NLP Analysis**

**1. Entity Recognition**

```python
def extract_entities(text):
    # base url where  HTTP requests will be sent for entity extraction
    url = "https://api.textrazor.com/"
    # defining http headers api_key to authenticate the request with the TextRazor service.
    headers = {
        "x-textrazor-key": api_key,
    }
    # defining the data on which we perform entity recognition analysis
    data = {
        "text": text,
        "extractors": "entities",
    }
    # sending http post request to the textRazor api url and respone from api stored in the response variable
    response = requests.post(url, headers=headers, data=data)
    # checks whether the api request was successful or not
    if response.status_code == 200:
        # converting the json response from the API into a Python dictionary
        result = response.json()
        # initialized 3 empty sets to store unique categorized values in respective sets
        drugs = set()
        diseases = set()
        study = set()
        # verifies whether the result dictionary contains the keys "response" and "entities".
        if "response" in result and "entities" in result["response"]:
            # iterates through each entity
            for entity in result["response"]["entities"]:
                # retrieves the list of entity types associated with the current entity and stores in entity_types
                entity_types = entity.get("type", [])
                # retrieve the name of current entity
                entity_name = entity.get("entityId", "")
                # checks whether the entity belongs to a specific entity_type
                # and stores in corresponding set if match found
                if "Drug" in entity_types:
                    drugs.add(entity_name)
                if "Disease" in entity_types:
                    diseases.add(entity_name)
                if "Study" in entity_types:
                    study.add(entity_name)
        # returns the dictionary of categorized entities
        return {
            "Drugs": ", ".join(sorted(drugs)),
            "Diseases": ", ".join(sorted(diseases)),
            "Study": ", ".join(sorted(study))
        }
    # if api request was not successfull prints error values
    else:
        print(f"Error: {response.status_code}, Response: {response.text}")
        return {
            "Drugs": "Error",
            "Diseases": "Error",
            "Study": "Error"
        }

# stores the test column(performing analysis on this column) in to text_column
text_column = combined_df['Combined']
# applying the function to each value of text_column
entities_data = text_column.apply(extract_entities)
# converts list of dictionaries into pandas dataframe
categorized_df = pd.DataFrame(entities_data.tolist())
# inserting the actual text column at first position in the categorized_df
categorized_df.insert(0, 'Text', combined_df['Combined'])
```

**Topic Classification and Sentiment Analysis**

```python
# importing textrazor for text analysis
import textrazor
# api_key for authenticating api requests
textrazor.api_key = "11c19b51ed0c70f07ba2dcadeed29fb0f5cb407659b220b724bd25f0"
# creating textrazor client with the ["entities", "topics", "sentences"] extractors
client = textrazor.TextRazor(extractors=["entities", "topics", "sentences"])
# topics_keywords is a dictionary, each key of dictionary mapped with a list of keywords
topics_keywords = {
    "Efficacy-General": ["efficacy", "effectiveness"],
    "Progression Free Survival (PFS)": ["progression-free survival", "PFS"],
    "Overall Survival (OS)": ["overall survival", "OS"],
    "Safety-General": ["safety", "safe"],
    "Safety-Side Effects": ["side effects", "adverse effects", "toxicity"],
    "General Opinion": ["opinion", "view", "thought"],
}
# function definition to classify the given text based on keywords
def classify_topics(text):
    # initialised an empty list to store detected topics
    detected_topics = []
    # iterates through each topic and its keywords
    for topic, keywords in topics_keywords.items():
        # converts the text to lowercase if keyword matches with current topic
        if any(keyword in text.lower() for keyword in keywords):
            # adds the topic to the list
            detected_topics.append(topic)
    # returns the detected_topics
    return detected_topics if detected_topics else ["Others"]

# function definition for the sentiment analysis
def analyze_sentiment(text):
    # if the text contains positive words returns positive sentiment
    if "good" in text.lower() or "positive" in text.lower() or "remarkable" in text.lower():
        return "Positive"
    # if the text contains negative words return negative sentiment
    elif "bad" in text.lower() or "negative" in text.lower() or "concerning" in text.lower():
        return "Negative"
    # if not both returns neutral sentiment
    else:
        return "Neutral"


# defines a function to process a column of text data
def process_text_column(text_column):
    # Initialized an empty list to store the each processed data
    results = []
    # iterates through each row of text_column
    for text in text_column:
        # stores the text data of each row while iterating
        row_result = {"Text": text}
        # sending text as parameter to the classify_topics and returned list stores in detected_topics
        detected_topics = classify_topics(text)
        # iterates through each topic in keys of topic_keywords
        for topic in topics_keywords.keys():
            # if topic matches in detected_topics, then analysing the sentiment of text
            if topic in detected_topics:
                sentiment = analyze_sentiment(text)
            else:
                sentiment = "Neutral"
            # each sentiment is get stored in respective topic column
            row_result[topic] = sentiment
        # if topic is others then analysing the sentiment of that text
        if "Others" in detected_topics:
            row_result["Others"] = analyze_sentiment(text)
        # if others is not in detected_topics, then assigning sentiment as neutral
        else:
            row_result["Others"] = "Neutral"
        # adding each row to the results from backwards
        results.append(row_result)
    return results
# stores the test column(performing analysis on this column) in to text_column
text_column = combined_df['Combined']
# starting the function call from process_text_column and ends with analyse_sentiment
processed_data = process_text_column(text_column)
# processed_data is list of dictionaries. converting it to pandas dataframe
task2_df = pd.DataFrame(processed_data)
task2_df.drop(columns=['Text'], inplace=True)
```
